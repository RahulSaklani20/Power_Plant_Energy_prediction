```python
#Defining the model.
def model_reg():
  model = Sequential()
  model.add(Dense(8, input_dim=4, kernel_initializer='uniform', activation='relu'))
  model.add(Dense(4, input_dim=4, kernel_initializer='uniform', activation='relu'))
  model.add(Dense(1, kernel_initializer='uniform', activation='linear'))

  opti = Adam(learning_rate=0.001)
  model.compile(loss='mean_squared_error', optimizer=opti, metrics=['mse'])
  return model
```

## Hyperparameter Tuning - learning Rate & Dropout Rate

```python
from keras.layers import Dropout
```

```python
def model_reg1(learning_rate,dropout_rate):
  model = Sequential()
  model.add(Dense(8, input_dim=4, kernel_initializer='uniform', activation='relu'))
  model.add(Dropout(dropout_rate))
  model.add(Dense(4,input_dim=4, kernel_initializer='uniform', activation='relu'))
  model.add(Dropout(dropout_rate))
  model.add(Dense(1, kernel_initializer='uniform', activation='linear'))

  opti = Adam(lr=learning_rate)
  model.compile(loss='mean_squared_error', optimizer=opti, metrics=['mse'] )
  return model
```

```python
model2 = KerasRegressor(build_fn=model_reg1, verbose=0, batch_size=50, epochs=100)
learning_rate = [0.001,0.01,0.1]
dropout_rate = [0.0,0.1,0.2]
param_grids = dict(learning_rate=learning_rate, dropout_rate=dropout_rate)
grid = GridSearchCV(estimator = model2,param_grid = param_grids,cv = KFold(),verbose = 10)
```

```python
gird_result1 = grid.fit(X_train, Y_train)
```

```python
print('Best : {}, using {}'.format(gird_result1.best_score_,gird_result1.best_params_))
means = gird_result1.cv_results_['mean_test_score']
stds = gird_result1.cv_results_['std_test_score']
params = gird_result1.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
  print('{},{} with: {}'.format(mean, stdev, param))
```

## Hyperparameter Tuning - Activation Function & Kernel Initializer

```python
# Defining the function
def model_reg2(activation_function,init):
  model = Sequential()
  model.add(Dense(8, input_dim=4, kernel_initializer=init, activation=activation_function))
  model.add(Dropout(0.0))
  model.add(Dense(4, input_dim=4, kernel_initializer=init, activation=activation_function))
  model.add(Dropout(0.0))
  model.add(Dense(1,activation='linear'))

  opti = Adam(learning_rate=0.001)
  model.compile(loss='mean_squared_error', optimizer=opti, metrics=['mse'])
  return model
```

```python
model3 = KerasRegressor(build_fn=model_reg2,verbose=0, batch_size=50, epochs=100)
activation_function = ['softmax','relu','tanh','linear']
init = ['uniform','normal','zero']
param_grids = dict(activation_function = activation_function,init = init)
grid = GridSearchCV(estimator = model3,param_grid = param_grids,cv = KFold(),verbose = 10)
```
